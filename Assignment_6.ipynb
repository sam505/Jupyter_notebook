# Using superstoreData, calculate the average Sales per Region within Year of Order Date (10 pts)

```python
superstoreData5 = superstoreData.reset_index()
superstoreData5['Year'] = superstoreData5["Order Date"].dt.year
superstoreData5 = superstoreData5.set_index(['Year','Region'])
superstoreData5 = superstoreData5.sum(level=['Year','Region']).reset_index()
```
